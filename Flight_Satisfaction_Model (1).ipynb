```python
def object_type_distribution_plot(df):
    """ This gives the distribution of the variables with the object data type
    """
    for var in cat_cols:

        print(df[var].describe(include='object'))
        plt.figure(figsize=(10,5))
        ax = sns.countplot(data=df, x=var)
        #rotate label if too long
        plt.xticks(rotation=90)
        plt.show()



def numeric_type_distribution_plot(df):
    """ This gives the distribution of the variables with the numeric data type
    """
    for var in numeric_cols:
        print(df[var].describe())
        plt.figure(figsize=(10,5))
        ax = sns.distplot(df[var])
        #rotate label if too long
        plt.xticks(rotation=90)
        plt.show()
```

#  Data Preprocessing

```python
# make data copy
df = data.copy()
```

```python
for col in cat_cols:
    print(col)
    print(df[col].unique())
```

```python
# Manual Label Encoding for categorical variables
satisfaction_mapper = {'satisfied':0, 'dissatisfied':1}
cust_type_mapper = {'Loyal Customer':0, 'disloyal Customer':1}
travel_type_mapper = {'Personal Travel':0, 'Business travel':1}
class_mapper  = {'Eco':0, 'Business':2, 'Eco Plus':1}

df["satisfaction"] = df["satisfaction"].map(satisfaction_mapper)
df["Customer Type"] = df["Customer Type"].map(cust_type_mapper)
df["Type of Travel"] = df["Type of Travel"].map(travel_type_mapper)
df["Class"] = df["Class"].map(class_mapper)
```

```python
# drop unprocessed and invalid colums (We will be dropping Arrival delay since it is correlated with Departure delay and also contained missing values)
df.drop(columns=["Arrival Delay in Minutes"], inplace=True)
```

# Separate Independent variables and Dependent variable

```python
# assign independent features to X
X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
X.head()
```

```python
y
```

```python
# do train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# data scaling (applying normalization)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_trained_scaled = X_train.copy()
X_trained_scaled[X_trained_scaled.columns] = scaler.transform(X_trained_scaled)
X_test_scaled = X_test.copy()
X_test_scaled[X_test_scaled.columns] = scaler.transform(X_test_scaled)
```

# Modeling

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(X_trained_scaled,y_train) # train the model
```

```python
test_pred = clf.predict(X_test_scaled)
```

```python
# evaluate performance using accuracy, f1-score 
from sklearn.metrics import accuracy_score, classification_report

acc_score = accuracy_score(test_pred, y_test)
class_rep = classification_report(test_pred, y_test)

print("Accuracy: ", acc_score)
print("Report:\n ", class_rep)
```

```python
class_pred = clf.predict(X_test_scaled) # class prediction
prob_pred = clf.predict_proba(X_test_scaled) # class probability
```

```python
from sklearn.multiclass import OneVsOneClassifier


# One Versus All (OvA)
model_ova = LogisticRegression(multi_class='ovr')
model_ova.fit(X_trained_scaled, y_train)
y_pred_ova = model_ova.predict(X_test_scaled)
y_pred_proba_ova = model_ova.predict_proba(X_test_scaled)

# One Versus One (OvO)
model_ovo = OneVsOneClassifier(LogisticRegression())
model_ovo.fit(X_trained_scaled, y_train)
y_pred_ovo = model_ovo.predict(X_test_scaled)
```

```python
# Displaying the predictions
print("OvA Predictions:", y_pred_ova[:5])
print("OvA Probabilities:\n", y_pred_proba_ova[:5])
print("OvO Predictions:", y_pred_ovo[:5])
```
